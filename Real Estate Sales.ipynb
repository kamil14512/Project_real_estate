```python
import pandas as pd 
```

```python
import numpy as py 
```

```python
import matplotlib.pyplot as plt
```

```python
# zaimportowanie źródłowego datasetu |
estate_data = pd.read_csv('/home/student/Downloads/Data/real_estate_sales_2001-2021_GL.csv')
```

```python
# wyświetlenie tytułów kolumn w datasecie 
display(estate_data.head())
```

```python
display(estate_data.info())
```

```python
# sprawdzenie czy kolumna "Serial Number" zawiera zduplikowane wartości
duplicates = estate_data['Serial Number'].duplicated()
display(duplicates)
```

```python
# usunięcie duplikujących się wartości z kolumny "Serial Number"
estate_data = estate_data.drop_duplicates(subset='Serial Number')
display(estate_data)
```

```python
# Kowersja danych w kolumnie 'Date Recorded' na format datetime
estate_data['Date Recorded'] = pd.to_datetime(estate_data['Date Recorded'])
```

```python
# Sortowanie dataframe po kolumnie 'Data Recorded'
estate_data = estate_data.sort_values(by='Date Recorded')
display(estate_data)
```

# Sprawdzenie w których latach zosatało sprzedane najwięcej nieruchomości

```python
# Dodanie nowej kolumny 'Year', która ekstraktuje tylko rok z kolumny 'Date Recorded'
estate_data['Year'] = estate_data['Date Recorded'].dt.year

# Grupowanie danych według roku i zliczanie liczby wystąpień
sales_per_year = estate_data['Year'].value_counts().sort_index()

# Sortowanie wartości od najmniejszych do największych

sales_per_year_sorted = sales_per_year.sort_values()

print(sales_per_year_sorted)

# Ustawienia wykresu
plt.figure(figsize=(10, 6))

# Tworzenie wykresu słupkowego
plt.bar(sales_per_year_sorted.index, sales_per_year_sorted.values, color='skyblue')

# Dodanie tytułu i etykiet
plt.title('Liczba sprzedanych mieszkań w poszczególnych latach')
plt.xlabel('Rok')
plt.ylabel('Liczba sprzedanych mieszkań')

# Dodanie siatki
plt.grid(axis='y', linestyle='--', linewidth=0.7, color='gray', alpha=0.7)

# Obrót etykiet na osi X
plt.xticks(rotation=45)

# Wyświetlenie wykresu
plt.show()
```

# W którym miastach zostało sprzedanych najwięcej i najmniej nieruchomości w wybranych latach  2010-2020 i przedstawienie tego na wykresie. 

```python
# Filtrowanie danych dla wybranych lat (2010-2020)
selected_years = estate_data[(estate_data['Year'] >= 2010) & (estate_data['Year'] <= 2020)]

# Grupowanie danych według miasta (kolumna 'Town') i zliczanie liczby sprzedanych nieruchomości
sales_per_town = selected_years['Town'].value_counts()

# Znalezienie miast z największą i najmniejszą liczbą sprzedaży
most_sales_town = sales_per_town.idxmax()
least_sales_town = sales_per_town.idxmin()

print(f'Miasto z największą liczbą sprzedaży: {most_sales_town} ({sales_per_town[most_sales_town]} sprzedaży)')
print(f'Miasto z najmniejszą liczbą sprzedaży: {least_sales_town} ({sales_per_town[least_sales_town]} sprzedaży)')

# Tworzenie wykresu
plt.figure(figsize=(12, 40))
sales_per_town_sorted = sales_per_town[sales_per_town > 0].sort_values(ascending=True)
sales_per_town_sorted.plot(kind='barh', color='skyblue')
plt.title('Liczba sprzedanych nieruchomości w poszczególnych miastach (2010-2020)')
plt.xlabel('Liczba sprzedanych nieruchomości')
plt.ylabel('Miasto')
plt.show()
```

# Przedstawienie jaka była średnia różnica pomiędzy wartości szacowana, a wartością sprzedaży w poszczególnych latach 

```python
# Obliczenie różnicy między 'Assessed Value' a 'Sale Amount'
estate_data['Difference'] = estate_data['Assessed Value'] - estate_data['Sale Amount']

# Grupowanie danych według roku i obliczenie średniej różnicy dla każdego roku
average_difference_per_year = estate_data.groupby('Year')['Difference'].mean()

# Tworzenie DataFrame z wynikami
result_df = pd.DataFrame({
    'Rok': average_difference_per_year.index,
    'Średnia różnica (USD)': average_difference_per_year.values
})

print(result_df)
```

```python
# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.bar(average_difference_per_year.index, average_difference_per_year.values, color='skyblue')

# Dodanie tytułu i etykiet osi
plt.title('Średnia różnica między wartością szacowaną a sprzedażową w poszczególnych latach')
plt.xlabel('Rok')
plt.ylabel('Średnia różnica (USD)')

# Pokazanie wykresu
plt.xticks(average_difference_per_year.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Sprawdzenie w którym mieście były sprzedawne nieruchomości o największej wartości, przedstawienie top 10 na wykresie 

```python
# Przekształcenie kolumny 'Sale Amount' na liczbę 
estate_data['Sale Amount'] = pd.to_numeric(estate_data['Sale Amount'], errors='coerce')

# Grupowanie danych według kolumny 'Town' i sumowanie 'Sale Amount'
sales_by_town = estate_data.groupby('Town')['Sale Amount'].sum().reset_index()


# Sortowanie danych według wartości sprzedaży malejąco
sales_by_town = sales_by_town.sort_values(by='Sale Amount', ascending=False)

# Wybranie top 10 miast
top_10_sales_by_town = sales_by_town.head(10)

# Konwersja nazw miast i wartości sprzedaży na listy
towns = top_10_sales_by_town['Town'].tolist()
sale_amounts = top_10_sales_by_town['Sale Amount'].tolist()

# Ustawienia wykresu
plt.figure(figsize=(12, 8))

# Tworzenie wykresu słupkowego
plt.barh(towns, sale_amounts, color='skyblue')

# Dodanie tytułu i etykiet
plt.title('Top 10 miast z najwyższą sumaryczną wartością sprzedaży nieruchomości')
plt.xlabel('Wartość sprzedaży (USD)')
plt.ylabel('Miasto')

# Odwrócenie osi Y, aby miasto z największą sprzedażą było na górze
plt.gca().invert_yaxis()

# Dodanie gridu 
plt.grid(axis='x', linestyle='--', linewidth=0.7, color='gray', alpha=0.7)

# Wyświetlenie wykresu
plt.show()
```

# Sprawdzenie jaka część nieruchomości została sprzedana w tym samym roku w który zostałą wystawiona 

```python
# Konwersja kolumny 'Date Recorded' na datę
estate_data['Date Recorded'] = pd.to_datetime(estate_data['Date Recorded'], errors='coerce')

# Ekstrakcja roku z kolumny 'Date Recorded'
estate_data['Year Recorded'] = estate_data['Date Recorded'].dt.year

# Sprawdzenie, czy nieruchomość została sprzedana w tym samym roku, w którym została wystawiona
estate_data['Same Year Sale'] = estate_data['List Year'] == estate_data['Year Recorded']

# Liczenie i obliczanie procentów
same_year_sales_counts = estate_data['Same Year Sale'].value_counts()
same_year_sales_percentage = estate_data['Same Year Sale'].value_counts(normalize=True) * 100

# Tworzenie tabeli wyników
results_table = pd.DataFrame({
    'Count': same_year_sales_counts,
    'Percentage': same_year_sales_percentage
})

# Zmiana indeksów na bardziej zrozumiałe
results_table.index = ['Same Year Sale', 'Different Year Sale']

print(results_table)
```

# Zaprezentowanie ile nieruchomości zostało sprzedanych w poszczególnych latach i jaka była ich średnia cena. Prezentacja zmiany ceny na przestrzeni lat w formie graficznej

```python
# Grupowanie danych według roku i obliczanie liczby sprzedanych nieruchomości oraz średniej ceny
yearly_sales_data = estate_data.groupby('Year Recorded').agg(
    Count=('Serial Number', 'count'),
    Average_Price=('Sale Amount', 'mean')
).reset_index()

# Zaprezentowanie tabeli wyników
print(yearly_sales_data)

# Wykres zmiany średniej ceny na przestrzeni lat
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales_data['Year Recorded'], yearly_sales_data['Average_Price'], marker='o', linestyle='-', color='b')

# Dodanie tytułu i etykiet
plt.title('Średnia cena sprzedaży nieruchomości na przestrzeni lat')
plt.xlabel('Rok')
plt.ylabel('Średnia cena sprzedaży (USD)')

# Dodanie siatki
plt.grid(True)

# Wyświetlenie wykresu
plt.show()
```
